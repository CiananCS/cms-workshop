```python
import os
import numpy

distance_file = os.path.join('data','data','distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float)
```

```python
import matplotlib.pyplot as plt
```

```python
sample = headers[1]

plt.figure()
fig1 = plt.plot(data[:,1],label=sample)
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()

plt.savefig('{}.png'.format(sample), dpi = 300)
```

```python
plt.figure()
for col in range(1, len(headers)):
    fig = plt.plot(data[:,col], label=headers[col])
    plt.legend()

plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.savefig('all_samples.png')
```

```python
#Exercise 1.2

#Finding the ymin and ymax
ymin = ymax = data[3,3] #random, doesn't matter as long as within theoretical ylim

for col in range(1, len(headers)):
    ymin = min(min(data[:,col]),ymin)
    ymax = max(max(data[:,col]),ymax)
print(ymin,ymax)
    
for col in range(1, len(headers)):
    plt.figure()
    fig = plt.plot(data[:,col], label=headers[col])
    plt.legend()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.ylim(ymin*0.9,ymax*1.1)
    

#plt.savefig('all_samples.png')
```
